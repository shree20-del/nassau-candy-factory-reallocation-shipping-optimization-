```python


df = pd.read_csv(r"C:\Users\dhans.SHREE\Desktop\Nassau-Candy-Analysis-v2-main\Nassau-Candy-Analysis-v2-main\data\cleaned_data.csv")
print("Original Shape:", df.shape)
```

```python
print("Before dropna:", df.shape)

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.dropna(subset=['Sales', 'Gross Profit', 'Units'])

print("After dropna:", df.shape)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True, errors='coerce')
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
```

```python
df['Gross Margin %'] = (df['Gross Profit'] / df['Sales']) * 100

df['Profit per Unit'] = df['Gross Profit'] / df['Units']

df['Revenue Contribution %'] = (df['Sales'] / df['Sales'].sum()) * 100

df['Profit Contribution %'] = (df['Gross Profit'] / df['Gross Profit'].sum()) * 100
```

```python
product_kpi = df.groupby('Product Name').agg({
    'Sales': 'sum',
    'Gross Profit': 'sum',
    'Units': 'sum',
    'Gross Margin %': 'mean'
}).reset_index()

# Gross Margin %
product_kpi['Gross Margin %'] = (
    product_kpi['Gross Profit'] / product_kpi['Sales']
) * 100

# Profit per Unit
product_kpi['Profit per Unit'] = (
    product_kpi['Gross Profit'] / product_kpi['Units']
)

# Revenue Contribution %
product_kpi['Revenue Contribution %'] = (
    product_kpi['Sales'] / product_kpi['Sales'].sum()
) * 100

# Profit Contribution %
product_kpi['Profit Contribution %'] = (
    product_kpi['Gross Profit'] / product_kpi['Gross Profit'].sum()
) * 100

print("Original Shape:", df.shape)
print("Original Shape:", product_kpi.shape)
```

```python
margin_volatility = df.groupby('Product Name')['Gross Margin %'].std().reset_index()

margin_volatility.columns = ['Product Name', 'Margin Volatility']
```

```python
product_kpi = product_kpi.merge(margin_volatility, on='Product Name', how='left')
```

```python
division_kpi = df.groupby('Division').agg({
    'Sales': 'sum',
    'Gross Profit': 'sum',
    'Units': 'sum',
    'Gross Margin %': 'mean'
}).reset_index()

df["Gross Margin %"] = (df["Gross Profit"] / df["Sales"]) * 100


division_kpi['Profit per Unit'] = (
    division_kpi['Gross Profit'] / division_kpi['Units']
)
print("Original Shape:", division_kpi.shape)
```
